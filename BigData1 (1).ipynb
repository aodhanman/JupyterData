```python
import pandas as pd

df = pd.read_csv('SchoolResultsFinal.csv')

# Calculate the mean number of students
mean_students = df['NumberStudents'].mean()

# Split the DataFrame into two based on the mean number of students
large_schools = df[df['NumberStudents'] > mean_students]
small_schools = df[df['NumberStudents'] <= mean_students]

# Calculate the mean MathMean, ReadingMean, and WritingMean for each DataFrame
large_schools_mean = large_schools[['MathMean', 'ReadingMean', 'WritingMean']].mean()
small_schools_mean = small_schools[['MathMean', 'ReadingMean', 'WritingMean']].mean()

# Print the details of the large schools
print('Large Schools:')
print(large_schools)

# Print the mean MathMean, ReadingMean, and WritingMean for the large schools
print('\nMean Scores for Large Schools:')
print(large_schools_mean)

# Print the details of the small schools
print('\nSmall Schools:')
print(small_schools)

# Print the mean MathMean, ReadingMean, and WritingMean for the small schools
print('\nMean Scores for Small Schools:')
print(small_schools_mean)
```

#question 1 iii

output the mean writing value for all schools in each district. 
Your code should print out a line for each district with the ID of the district (DBN) 
and the average writing grade achieved by the schools (collectively) in that district

```python
import pandas as pd

df = pd.read_csv('SchoolResultsFinal.csv')

district_mean_writing = df.groupby('DBN')['WritingMean'].mean()

# Print the grade for each district
for dbn, mean_writing in district_mean_writing.items():
    print(f'District {dbn}: {mean_writing:.2f}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SchoolResultsFinal.csv')

# Calculate the grade for each district
district_mean_writing = df.groupby('DBN')['WritingMean'].mean()

district_mean_writing.plot(kind='bar', figsize=(12,6))
plt.xlabel('DBN')
plt.show()

# shwow bar graph of the writing grade per district

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SchoolResultsFinal.csv')

# mean grade for each subject by district
district_mean = df.groupby('DBN')[['WritingMean', 'ReadingMean',  'MathMean']].mean()

# Line graph for each district
district_mean.plot(kind='line', figsize=(12,6))
plt.title('Mean Grade by District')
plt.xlabel('District')
plt.ylabel('Mean Grade')
plt.xticks(range(1, 33))
plt.legend(['Writing', 'Reading',  'Maths'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SchoolResultsFinal.csv')

# Create a scatter plot of MathMean vs. ReadingMean
plt.scatter(df['MathMean'], df['ReadingMean'])
plt.title('MathMean vs. ReadingMean')
plt.xlabel('MathMean')
plt.ylabel('ReadingMean')
plt.show()

# Create a scatter plot of MathMean vs. WritingMean
plt.scatter(df['MathMean'], df['WritingMean'])
plt.title('MathMean vs. WritingMean')
plt.xlabel('MathMean')
plt.ylabel('WritingMean')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('SchoolResultsFinal.csv')

# missing values in the WritingMean column
num_missing = df['WritingMean'].isnull().sum()
print('Number of missing values in WritingMean:', num_missing)

# replace all missing values in the ‘WritingMean’ column with the median value of the ‘WritingMean’ column. 
median_writing = df['WritingMean'].median()
df['WritingMean'].fillna(median_writing, inplace=True)

# check for missing values in the WritingMean column
num_missing = df['WritingMean'].isnull().sum()
print('Number of missing values in WritingMean after replacement:', num_missing)
```
